## Transformar a matriz em uma matriz triangular

O processo de transformar a matriz em uma matriz triangular é essencial, pois facilita o calculo computacional.
Neste sentido matrizes triangulares são aquelas que seguem o seguinte formato:

$$
M =
\begin{bmatrix}
a & b  \\
0 & c \\
\end{bmatrix}
$$

```python
# transformação da lista em uma matriz pela função array
M_2x2 = array(M_2x2, dtype = float)

#Nova matriz para que não haja perca de informações
matriz_copia_2x2 = M_2x2.copy()
```

Antes de qualquer passo, verificaremos se a primeira coluna da matriz é zero
ou se o elemento 1x1 é zero.

Apos isso será necessario dividir a primeira linha da matriz pelo seu primeiro elemento.
E depois subtrir a linha 2 pela linha 1, para que tenhamos o primeiro 
digito da segunda linha como 0.

```python
# para o calculo da determinante sera armazenado a variavel troca
troca_2x2 = 0


# Caso toda a primeira coluna seja composta por zeros nada acontece
if all(linha[0] == 0 for linha in matriz_copia_2x2):

    linha1 = None
    
else:
    # Caso o primeiro elemento da matriz seja 0, realizar uma troca das linhas por um valor nâo nulo    
    if matriz_copia_2x2[0][0] == 0:

        for i in range(1,2):

            if matriz_copia_2x2[i][0] != 0:

                matriz_copia_2x2[[0,i]] = matriz_copia_2x2[[i,0]]
                troca_2x2 += 1
                break
            

    # Armazenar o elemento 1x1 original
    elemento_original_1x1_da_matriz_2x2 = matriz_copia_2x2[0][0]

    # Dividir todos os elementos da primeira linha pelo elemento 1x1 da matriz: Assim o elemento 1x1 = 1         
    linha1 = matriz_copia_2x2[0]/matriz_copia_2x2[0][0]

    # Substituir a linha 1  subtrair a linha 2 pela linha 1 multiplicada pelo elemento 2x1
    matriz_copia_2x2[0] = linha1
    
    #visualize
    print("Matriz antes da subtração:")
    print(matriz_copia_2x2)
    
    # subtrair a linha 2 pela linha 1 multiplicada pelo elemento 2x1
    matriz_copia_2x2[1]= matriz_copia_2x2[1] - matriz_copia_2x2[0]*matriz_copia_2x2[1][0]
    
#Armazenar o elemento caso a coluna seja composta por zeros
if matriz_copia_2x2[0][0] == 0:
    elemento_original_1x1_da_matriz_2x2 = matriz_copia_2x2[0][0]

#visualize
print("")
print("Matriz de após a subtração")
print(matriz_copia_2x2)
```

**Portanto agora obtivemos uma matriz triangular 2x2 (desconsiderando os resultados do sistema)**

## Descobir a determinante da matriz 2x2
O calculo da determinante é realizado pela multiplicação da diagonal principal vezes -1 elevado ao numero de trocas de linhas

```python
def determinante_2x2(M):
    """
    Arg: Matriz triangular 2x2
    
    Essa função apenas funciona para matrizes triangulares 2x2, 
    visto que a ultima coluna 
    corrsponde as respostas do sistema
    
    Return: um float do determinante de matriz 
    """
    determinante = M[0][0]
    for linha in range (1, len(M)):
        determinante *= M[linha][linha]
        
    #Compensamos a divisao realizada anteriormente para criar a matriz triangular
    determinante = determinante * elemento_original_1x1_da_matriz_2x2 * ((-1)**troca_2x2)
    
    return determinante
```

```python
#visualize
print(f"sua determinante é {determinante_2x2(matriz_copia_2x2)}")
```

## Descobir se o sistema 2x2 é possivel determinado, possivel indeterminado ou impossivel

```python
def tipo_sistema_2x2(M):
    '''
    Arg: Uma matriz triangular 2X2
    
    Contem os mesmos requisistos para o calculo da determinante
    
    Return: Uma string que expressa o tipo do sistema
    
    
    
    '''
    if determinante_2x2(M) == 0:
        # Caso uma linha seja compostas apenas por zeros o sistema é indeterminado
        if all(x == 0 for x in M[0]):
            return "Seu sistema é possivel indeterminado"              
        elif all(x == 0 for x in M[1]):
            return "Seu sistema é possivel indeterminado"   
            
        # Caso não, ele é impossivel
        else:
            return "Seu sistema é impossivel"
    else:
        return "Seu sistema é possivel determinado"
```

```python
print(tipo_sistema_2x2(matriz_copia_2x2))
```

## Descobir as soluções do sistema 2x2

```python
def respostas_sistema_2x2(M):
    '''
    Arg: Matriz triangular 3X3 
    As mesmas restrições que se aplicam as funções anterores
    
    Retrun: Dicionario contendo as respostas 
    caso a função seja possivel determinada
    Ou None caso não haja soluções
    
    '''

    
    if tipo_sistema_2x2(M) == "Seu sistema é possivel indeterminado":
        return None
    elif tipo_sistema_2x2(M) == "Seu sistema é impossivel":
        return None
    else:
        respostas = {}
        
        y = M[1][2]/ M[1][1]
        x = (M[0][2] - M[0][1]*y ) / M[0][0]
        
        respostas["x"], respostas["y"] = x,y
        return respostas
```

```python
#visualize
print(respostas_sistema_2x2(matriz_copia_2x2))
```

## Sumario do Sistema linear 2x2
Infromações gerais sobre o sistema

```python
"""
Exibe um resumo final do sistema.

Mostra:
sistema original
determinante
classificação
soluções (caso existam)
"""

print(f"Seu sistema inicial é:\n{M_2x2}")
print("")
print(f"A determinante da matriz é {determinante_2x2(matriz_copia_2x2)}")

print("")
print(tipo_sistema_2x2(matriz_copia_2x2))
print("")

if respostas_sistema_2x2(matriz_copia_2x2) is not None:

    print(f"Suas variáveis x, y e z são respectivamente:")

    print( f"x = {respostas_sistema_2x2(matriz_copia_2x2)['x']}" )

    print(f"y = {respostas_sistema_2x2(matriz_copia_2x2)['y']}")
else:
    print("Não foram encontradas soluções ao seu sistema")
```

Neste sentido matrizes triangulares são aquelas que seguem o seguinte formato:

$$
M =
\begin{bmatrix}

0 & d & e \\

```python
# transformação da lista em uma matriz pela função array
M_3x3 = array(M_3x3, dtype = float)

# Nova matriz para que não haja perca de informações
matriz_copia_3x3 = M_3x3.copy()
```


Antes de qualquer passo, verificaremos se a primeira coluna da matriz é zero
ou se o elemento 1x1 é zero.

Apos isso será necessario dividir a primeira linha da matriz pelo seu primeiro elemento.
E depois subtrir a linha 2 pela linha 1, para que tenhamos o primeiro 
digito da segunda linha como 0.



```python
#para o calculo da determinante sera armazenado a variavel troca
troca_3x3 = 0

# Caso toda a primeira coluna seja composta por zeros nada acontece
if all(linha[0] == 0 for linha in matriz_copia_3x3):

    linha1 = None
    
else:
    # Caso o primeiro elemento da matriz seja 0, realizar uma troca das linhas até um valor nâo nulo    
    if matriz_copia_3x3[0][0] == 0:

        for i in range(1,3):

            if matriz_copia_3x3[i][0] != 0:

                matriz_copia_3x3[[0,i]] = matriz_copia_3x3[[i,0]]
                troca_3x3 += 1
                break
            

    # Armazenar o elemento 1x1 original
    elemento_original_1x1_da_natriz_3x3 = matriz_copia_3x3[0][0]

    # Dividir todos os elementos da primeira linha pelo elemento 1x1 da matriz: Assim o elemento 1x1 = 1         
    linha1 = matriz_copia_3x3[0]/matriz_copia_3x3[0][0]

    # Substituir a linha 1  subtrair a linha 2 pela linha 1 multiplicada pelo elemento 2x1
    matriz_copia_3x3[0] = linha1
    
    #visualize
    print("Matriz antes da subtração:")
    print(matriz_copia_3x3)

    # subtrair a linha 2 pela linha 1 multiplicada pelo elemento 2x1
    matriz_copia_3x3[1]= matriz_copia_3x3[1] - matriz_copia_3x3[0]*matriz_copia_3x3[1][0]

#Armazenar o elemento caso a coluna seja composta por zeros
if matriz_copia_3x3[0][0] == 0:
    elemento_original_1x1_da_natriz_3x3 = matriz_copia_3x3[0][0]

#visualize
print("")
print("Matriz de após a subtração")
print(matriz_copia_3x3)
```

```python
'''
Agora tornaremos 0 o elemento 3x1 da matriz
'''
#subtrair a linha 3 da matriz pela linha 1 multiplicada pelo elemento 3x1
matriz_copia_3x3[2] = matriz_copia_3x3[2] - matriz_copia_3x3[0]*matriz_copia_3x3[2][0]

#visualize
print("Seu sistema após a subtração")
print(matriz_copia_3x3)
```

## Descobir a determinante da matriz 3x3
O calculo da determinante é realizado pela multiplicação da diagonal principal vezes -1 elevado ao numero de trocas de linhas

```python
def determinante_3x3(M):
    """
    Arg: Matriz triangular
    
    Essa função apenas funciona para matrizes triangulares 2x2, 
    visto que a ultima coluna 
    corrsponde as respostas do sistema
    
    Return: um float do determinante de matriz 
    """
    determinante = M[0][0]
    for linha in range (1, len(M)):
        determinante *= M[linha][linha]
        
    #Compensamos a divisao realizada anteriormente para criar a matriz triangular
    determinante = determinante * elemento_original_1x1_da_natriz_3x3 * elemento_original_2x2_da_matriz_3x3 * ((-1)**troca_3x3)
    
    return determinante
```

## Descobir se o sistema 3x3 é possivel determinado, possivel indeterminado ou impossivel

```python
def tipo_sistema_3x3(M):
    '''
    Arg: Uma matriz triangular 3X3
    
    Contem os mesmos requisistos para o calculo da determinante
    
    Return: Uma string que expressa o tipo do sistema
    
    
    
    '''
    if determinante_3x3(M) == 0:
        #Caso uma linha seja compostas apenas por zeros o sistema é indeterminado
        if all(x == 0 for x in M[0]):
            return "Seu sistema é possivel indeterminado"  
            
        elif all(x == 0 for x in M[1]):
            return "Seu sistema é possivel indeterminado"
            
        elif all(x == 0 for x in M[2]):
            return "Seu sistema é possivel indeterminado"
        # Caso nãom é impossivel
        else:
            return "Seu sistema é impossivel"
    else:
        return "Seu sistema é possivel determinado"
    
```

## Descobir as soluções do sistema 3x3

```python
def respostas_sistema_3x3(M):
    '''
    Arg: Matriz triangular 3X3 
    As mesmas restrições que se aplicam as funções anterores
    
    Retrun: Dicionario contendo as respostas 
    caso a função seja possivel determinada
    Ou None caso não haja soluções
    
    '''

    
    if tipo_sistema_3x3(M) == "Seu sistema é possivel indeterminado":
        return None
    elif tipo_sistema_3x3(M) == "Seu sistema é impossivel":
        return None
    else:
        respostas = {}
        z = M[2][3] / M[2][2]
        y = (M[1][3] - M[1][2]*z) / M[1][1]
        x = (M[0][3] - M[0][1]*y - M[0][2]*z) / M[0][0]
        
        respostas["x"], respostas["y"], respostas["z"] = x,y,z
        return respostas
```
