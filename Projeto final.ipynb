O sistema linearar como por exemplo:

Caso se deseje se utilizar de um sistema 2X2 como:

```python
'''
Este será o formato adotado para a representação de sistemas e, consequentemente, matrizes neste progarama
onde cada linha da matriz é uma lista dentro de uma lista principal

Ou seja, o valor do elemento 1x1 da matriz pode ser acessado como A[0][0]
'''
M = [
    [2, 1, -1, 3],
    [1, -1, 2, 3],
    [3, 2, 1, 10]  
]
```

## Transformar a matriz em uma matriz triangular

O processo de transformar a matriz em uma matriz triangular é essencial, pois facilita o calculo computacional.
Neste sentido matrizes triangulares são aquelas que seguem o seguinte formato:

                                                    a  b  c

                                                    0  d  e 

                                                    0  0  f 

#### será usado a função array do  modulo numpy para transformar as listas em uma matriz

```python
from numpy import array
```

```python
# transformação e copia da matriz original para que nao haja perdas de informação
M = array(M, dtype = float)
matriz = array(M, dtype=float)
matriz_copia = matriz.copy()
```

```python
'''
Antes de qualquer passo, tranformaremos o primeiro digito da primeira linha em 1.
Para essa tranformação, será necessario dividir as linhas da matriz pelo primeiro elemento.
E depois subtrir as linhas entre si, para que o determinante dela se matenha o mesmo, e tenhamos o primeiro 
digito da segunda linha como 0.

'''
#Armazenar o elemento 1x1 original
elemento_original1 = matriz_copia[0][0]
# Dividir todos os elementos da primeira linha pelo elemento 1x1 da matriz: Assim o elemento 1x1 = 1
linha1 = matriz_copia[0]/matriz_copia[0][0]

#Substituir a linha 1 e subtrair a linha 2 pela linha 1 multiplicada pelo elemento 2x1
matriz_copia[0] = linha1
matriz_copia[1]= matriz_copia[1] - matriz_copia[0]*matriz_copia[1][0]
```

```python
#subtrair a linha 3 da matriz pela linha 1 multiplicada pelo elemento 3x1
matriz_copia[2] = matriz_copia[2] - matriz_copia[0]*matriz_copia[2][0]

#visualize
print(matriz_copia)
```

```python
'''
Da mesma forma, tornaremos 1 o segundo termo da segunda fileira (2x2)
fileira, dividindo a linha por ele mesmo, como fizemos anteriormente.
E subtrairemos a linha 3 pela linha 2.

'''
#Armazenar o elemento 2x2
elemento_original2 = matriz_copia[1][1]
#Dividir todos os elementos da segunda linha pelo elemento 2x2 da matriz: Assim o elemento 2x2 = 1
linha2 = matriz_copia[1]/matriz_copia[1][1]


#substituir a linha 2
matriz_copia[1] = linha2

#visualize
print(matriz_copia)
print("")

#subtrair a linha 3 pela linha 2 multiplicada pelo elemento 3x2
matriz_copia[2] = matriz_copia[2] - matriz_copia[1]*matriz_copia[2][1]

#visualize
print(matriz_copia)
```

## Descobir se a função é possivel determinada, possivel indeterminada ou impossivel

```python
def determinante_3x3(M):
    """
    Arg: Matriz triangular
    
    Essa função apenas funciona para matrizes triangulas 3x3, 
    visto que a ultima coluna 
    corrsponde as respostas do sistema
    
    Return: um float do determinante de matriz 
    """
    determinante = M[0][0]
    for linha in range (1, len(M)):
        determinante *= M[linha][linha]
        
    #Compensamos a divisao realizada anteriormente para criar a matriz triangular
    determinante = determinante * elemento_original1 * elemento_original2
    return determinante
```

```python
#visualize
print(determinante_3x3(matriz_copia))
```

```python
def tipo_sistema_3x3(M):
    '''
    Arg: Uma matriz triangular 3X3
    Contem os mesmos requisistos para o calculo da determinante
    Return: Uma string que expressa o tipo do sistema
    
    
    
    '''
    if determinante_3x3(M) == 0:
        if sum(M[0]) or sum(M[1]) or sum(M[2]) == 0:
            return "Seu sistema é possivel indeterminado"
        else:
            return "Seu sistema é impossivel"
    else:
        return "Seu sistema é possivel determinado"
    
```

```python
#visualize
print(tipo_sistema_3x3(matriz_copia))
print(matriz_copia)
```

```python
def respostas_sistema_3x3(M):
    '''
    Arg: Matriz triangular 3X3 
    As mesmas restrições que se aplicam as funções anterores
    Retrun: Dicionario contendo as respostas 
    caso a função seja possivel determinada
    
    '''

    
    if tipo_sistema_3x3(M) == "Seu sistema é possivel indeterminado":
        print("Existem infinitas soluções para o seu problema")
    elif tipo_sistema_3x3(M) == "Seu sistema é impossivel":
        print("Suas soluções não são possiveis de serem determinadas")
    else:
        respostas = {}
        z = M[2][3] / M[2][2]
        y = (M[1][3] - M[1][2]*z) / M[1][1]
        x = (M[0][3] - M[0][1]*y - M[0][2]*z) / M[0][0]
        
        respostas["x"], respostas["y"], respostas["z"] = x,y,z
        return respostas
```

```python
#visualize
respostas_sistema_3x3(matriz_copia)
```
