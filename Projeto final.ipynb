## Matriz de teste

```python
'''
Este será o formato adotado para a representação de matrizes neste progarama
onde cada linha da matriz é uma lista dentro de uma lista principal

Ou seja, o valor do elemento 1x1 da matriz pode ser acessado como A[0][0]
'''
A = [
    [1,2,3],
    [3,2,1],
    [2,3,1]  
]
```

#### A função Fractions sera utilizada pois permite uma notação mais precisa

```python
# permite que uma fração como 1/2 seja escrita como Fractions (1, 2) e não como 0.5
from fractions import Fraction
```

## Transformar a matriz em uma matriz triangular

O processo de transformar a matriz em uma matriz triangular é essencial, pois facilita o calculo computacional.
Neste sentido matrizes triangulares são aquelas que seguem o seguinte formato:

                                                    a  b  c

                                                    0  d  e 

                                                    0  0  f 

```python
#copia da matriz original para que nao haja perdas de informação
A1 = A.copy()
```

```python
'''
Antes de qualquer passo, tranformaremos o primeiro digito das duas primeiras linhas em 1.
Para essa tranformação, será necessario dividir as linhas da matriz pelo primeiro digito.
E depois subtrir as linhas entre si, para que o determinante dela se matenha o mesmo, e tenhamos o primeiro 
digito da segunda lnha como 0

'''
# Dividir todos os elementos da primeira linha pelo elemento 1x1 da matriz: Assim o elemento 1x1 = 1
linha1 = list(map(lambda x: Fraction(x, A[0][0]), A[0]))

# Dividir todos os elementos da segunda linha pelo elemento 2x1 da matriz: Assim o elemento 2x1 = 1
linha2 = list(map(lambda x: Fraction(x, A[1][0]), A[1]))


#Substituir a linha 1 e subtrair a linha 2 pela linha 1
A1[0], A1[1]= linha1, [a - b for a, b in zip(linha2, linha1)]
```

```python
#Confira seus resultados e sua nova matriz
print(linha1)
print("")
print(linha2)
print("")
print(A1)
print("")
```

```python
#Dividir todos os elementos da tereira linha pelo elemento 3x1 da matriz: Assim o elemento 3x1 = 1
linha3 = list(map(lambda x: Fraction(x, A[2][0]), A[2]))

#substiruir a linha 3
A1[2] = linha3


#subtrair a linha 3 pela linha 1
A1[2] = [a - b for a, b in zip(linha3, linha1)]
```

```python
#visualize
print(linha3)
print(A1)
```

```python
'''
Da mesma forma, tornaremos 1 o segundo termo da terceira (3x2)
fileira, dividindo a linha por ele mesmo, como fizemos anteriormente.
E subtrairemos pela linha 1 novamente.

'''
#Dividir todos os elementos da tereira linha pelo elemento 3x2 da matriz: Assim o elemento 3x2 = 1
linha3 = list(map(lambda x: Fraction(x) / Fraction(A[2][1]), A[2]))
print(linha3)

```
